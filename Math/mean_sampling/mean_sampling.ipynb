```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

Part c)

```python
np.random.seed(1)
```

```python
k = 5
n = 1000
mean_stupid = np.zeros(n)
mean_sample = np.zeros(n)
var = np.zeros(n)

for j in range(n):
    X = np.random.normal(0,1,k)
    mean_sample[j] = np.mean(X)
    mean_stupid[j] = np.mean(X[0:3])
    for i in range(k):
        var[j] = var[j] + (1/k)*(X[i]-mean_sample[j])**2
    #instead of the second for loop, you can also use
    #var[j] = np.var(X,ddof=1)
    #print(var[j], 'computation by hand')
    #print(np.var(X,ddof=1), 'numpy')
        
#print(mean)
#print(var)
```

Part d) - Plotting the 1000 values of the sample mean:

```python
plt.plot(mean_sample,'.', label = 'sample mean', color = 'red')
plt.plot(mean_stupid,'.', label = 'stupid mean', color='black')
plt.plot(np.zeros(n),label = 'true mean',linewidth = 2)
#plt.plot(np.ones(n)*np.mean(mean))
plt.legend()
```

There seems to be no bias, but the black dots seem to have larger variance:

```python
print(np.var(mean_sample))
print(np.var(mean_stupid))

```

Part e) - Plot the 1000 estimators for the variance

```python
plt.plot(var,'.')
plt.plot(np.ones(n),label = 'true var',linewidth = 2)
print(np.mean(var))
plt.legend()
```

There seems to be a bias: the average of sample variances lies below of the true variance.
